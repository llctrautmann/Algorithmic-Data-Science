```python
import numpy as np
import re
```

```python
x = "hello my name is luca and I am an avid student of artificial intelligence and algorithms"

a = 'The next question from researchers is when you will be able to show up at work'
b = 'That would be a great opportunity to take your technology build software and hardware, demonstrate at conferences and exhibitions.'
c = 'In fact its very interesting that a few of the top scientists of today know about the future of robotics'

strings = [a,b,c]
```

```python
def shingle(text: str, k: int):
    shingle_set = []
    
    for i in range(len(text)):
        shingle_set.append(text[i:i+k])
        
    return set(shingle_set)

def zero_vectoring(vocab,shingle_set,zero_v): 
    for i in range(len(vocab)):
        if list(vocab)[i] in shingle_set:
            zero_v[i] = 1
```

```python
k = 2
shingle_set = shingle(x,k)

a =shingle(a,k)
b =shingle(b,k)
c =shingle(c,k)
```

```python
vocab = list(a.union(b).union(c))
```

```python
vocab
```

```python
a_zero_v = np.zeros(len(vocab)).reshape(-1,1)
b_zero_v = np.zeros(len(vocab)).reshape(-1,1)
c_zero_v = np.zeros(len(vocab)).reshape(-1,1)

zero_vectors = [a_zero_v,b_zero_v,c_zero_v]
```

```python
for i in range(len(zero_vectors)):
    zero_vectoring(vocab,strings[i],zero_vectors[i])
```
